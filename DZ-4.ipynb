## Задание 1

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
boston.data.shape
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
X.info()
```

```python
y = pd.DataFrame(boston.target, columns=['Price'])
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

## Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
rfr.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = rfr.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
''' RandomForestRegressor показал лучший результат'''
```

## Задание 4

```python
cards = pd.read_csv('creditcard.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = cards.drop('Class', axis=1)
y = cards['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
y.value_counts(normalize=True)
```

```python
X_train.info()
```

```python
pd.options.display.max_columns = 100
```

```python
cards.head(10)
```

```python
cards.shape
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15], 
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100), 
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_predict_proba = clf.predict_proba(X_test)
y_predict_proba = y_predict_proba[:, 1]
y_predict_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_predict_proba)
```
